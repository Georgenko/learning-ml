```python
import numpy as np

from sklearn.linear_model import LogisticRegression
from utils import load_preprocessed_ds
```

```python
X_tr, y_tr, X_te, y_te, X_tr_pd = load_preprocessed_ds()
```

```python
# Model params
num_features = X_tr_pd.shape[1]
num_samples = X_tr_pd.shape[0]
W = np.random.randn(num_features)
b = np.random.randn()
```

```python
# Utils
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def forward_pass(inp, W, b):
    z = np.matmul(inp, W) + b
    a = sigmoid(z)
    return a

def loss_binary_cross_entropy(yh, y):
    loss_vector = -(y * np.log(yh) + (1 - y) * np.log(1 - yh))
    return np.mean(loss_vector)

def calculate_gradients(yh, y_tr, X_tr, num_samples):
    error = yh - y_tr
    dl_dw = (1 / num_samples) * np.matmul(X_tr.T, error)
    dl_db = (1 / num_samples) * np.sum(error)
    return dl_dw, dl_db
```

```python
# Training loop
epochs = 15000
lr = 0.1
for epoch in range(epochs):

    yh = forward_pass(X_tr, W, b)

    loss = loss_binary_cross_entropy(yh, y_tr)
    if epoch % 1000 == 0:
        print(f"Epoch: {epoch}. Loss: {loss}")

    dl_dw, dl_db = calculate_gradients(yh, y_tr, X_tr, num_samples)

    # update
    W -= lr * dl_dw
    b -= lr * dl_db
```

```python
# Check that it matches LogisticRegression from sklearn without regularization (penaly=None)
clf = LogisticRegression(penalty=None).fit(X_tr, y_tr)
print(loss_binary_cross_entropy(clf.predict_proba(X_tr)[:,1], y_tr))
```

```python
# Check accuracy against test dataset
yh = forward_pass(X_te, W, b)

yh_thresh = (yh > 0.5).astype(int)

accuracy = (yh_thresh == y_te).mean()
print(accuracy)
```
