```python
import pickle
import numpy as np
```

```python
# Load preprocessed dataset
with open("heart_disease_processed.pkl", "rb") as f:
    X_train, X_test, y_train, y_test = pickle.load(f)
```

```python
# Model params
num_features = X_train.shape[1]
num_samples = X_train.shape[0]
W = np.random.randn(num_features)
b = np.random.randn()

X_tr = X_train.to_numpy()
y_tr = y_train.to_numpy()
x_te = X_test.to_numpy()
y_te = y_test.to_numpy()
```

```python
# Utils
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def forward_pass(inp, W, b):
    z = np.matmul(inp, W) + b
    a = sigmoid(z)
    return a

def loss_binary_cross_entropy(yh, y):
    loss_vector = -(y * np.log(yh) + (1 - y) * np.log(1 - yh))
    return np.mean(loss_vector)

def calculate_gradients(yh, y_tr, X_tr, num_samples):
    error = yh - y_tr
    dl_dw = (1 / num_samples) * np.matmul(X_tr.T, error)
    dl_db = (1 / num_samples) * np.sum(error)
    return dl_dw, dl_db
```

```python
# Training loop
epochs = 1000
lr = 0.1
for epoch in range(epochs):

    yh = forward_pass(X_tr, W, b)

    loss = loss_binary_cross_entropy(yh, y_tr)
    if epoch % 50 == 0:
        print(f"Epoch: {epoch}. Loss: {loss}")

    dl_dw, dl_db = calculate_gradients(yh, y_tr, X_tr, num_samples)

    # update
    W -= lr * dl_dw
    b -= lr * dl_db
```
